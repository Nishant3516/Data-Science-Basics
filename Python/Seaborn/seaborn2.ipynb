# Seaborn
## Boxplot, Violinplot, Countplot, Barplot

```python
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

### Countplot

```python
sns.countplot(x='sex', data=df, hue='sex')
```

```python
sns.countplot(x='sex', data=df, hue='smoker')
```

```python
sns.countplot(y='sex', data=df, hue='smoker')
```

### Bar plot

```python
sns.barplot(x="day", y="total_bill", data=df, hue='sex')
```

### Box plot

```python
sns.boxplot(x='sex', y='total_bill', data=df, hue="sex")
```

### Violin Plot

```python
sns.violinplot(x='total_bill', y='day', data=df)
```
