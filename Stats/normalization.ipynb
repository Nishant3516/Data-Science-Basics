# Normalization (Min-Max Normalization)
- Scales down value in the range between 0 and 1 (both inclusive).
- Xnormalized = X – Xmin / Xmax – Xmin

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
                 header=None, usecols=[0, 1, 2])
df.columns = ['Class', 'Alcohol', 'Malic']
df.head()
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit_transform(df[['Alcohol', 'Malic']])
```
