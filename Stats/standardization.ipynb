# Standardization (z-score)
values scaled such that Mean = 0, sd = 1.
z = (x – μ) / σ

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
                 header=None, usecols=[0, 1, 2])
df.columns = ['Class', 'Alcohol', 'Malic']
df.head()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(df[['Alcohol', 'Malic']])
```
